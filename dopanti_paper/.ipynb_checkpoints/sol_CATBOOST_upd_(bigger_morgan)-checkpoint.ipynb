```python
import gc
import os
import pickle
import random
import joblib
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import average_precision_score as APS
from sklearn.model_selection import ShuffleSplit, StratifiedShuffleSplit
import duckdb
import pandas as pd
import mapply
from sklearn.metrics import average_precision_score, precision_recall_curve
from tokenizers import Tokenizer
from transformers import PreTrainedTokenizerFast
from transformers import AutoConfig, AutoTokenizer, AutoModel, DataCollatorWithPadding
import torch
import torch.nn as nn
import torch.nn.functional as F
import importlib
from torch.optim import Adam, AdamW
import matplotlib.pyplot as plt
import gc
import matplotlib as mpl
import json
from sklearn.metrics import classification_report, cohen_kappa_score
from sklearn.utils import class_weight
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
from sklearn.preprocessing import StandardScaler
# from keras.layers import Layer
import numpy as np
from catboost import CatBoostClassifier
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit import DataStructs
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
from sklearn.model_selection import train_test_split
# from tensorflow.keras.layers import LeakyReLU
class Wrapper:
    def __init__(self, method_name, module_name):
        self.method_name = method_name
        self.module = importlib.import_module(module_name)

    @property
    def method(self):
        return getattr(self.module, self.method_name)

    def __call__(self, *args, **kwargs):
        return self.method(*args, **kwargs)
        
wrapped_mol_from_smiles = Wrapper("MolFromSmiles", "rdkit.Chem")

mapply.init(
    n_workers=10,#was 5 before, trying 10,
    progressbar=True,
)

wrapped_has_substract_match = Wrapper("HasSubstructMatch", "rdkit.Chem")

# calc some descriptors
def calc_descriptors(smile):
    mol = wrapped_mol_from_smiles(smile)
    descr=[round(Descriptors.MolLogP(mol),4),
                  round(Descriptors.TPSA(mol),4),
                  Descriptors.NHOHCount(mol),
                  Descriptors.NOCount(mol),
                  Descriptors.NumHAcceptors(mol),
                  Descriptors.NumHDonors(mol),
                  Descriptors.NumRotatableBonds(mol),
                  Descriptors.NumHeteroatoms(mol),
                  round(Descriptors.FractionCSP3(mol),4),
                  round(Descriptors.ExactMolWt(mol),4),
                  rdMolDescriptors.CalcNumAromaticRings(mol)]
    return np.array(descr,dtype=np.float16)

def get_unique_values(list_of_lists):
    unique_values = []
    for inner_list in list_of_lists:
        for value in inner_list:
            unique_values.append(value)
    return unique_values
```

```python
def rdkit_numpy_convert(fp):
    output = []
    for f in fp:
        try:
            arr = np.zeros((1,))
            DataStructs.ConvertToNumpyArray(f, arr)
            arr2=arr
            output.append(arr)
        except:
            output.append(arr2)
            pass
    return np.asarray(output)
fp=[]
for m in tqdm(all_data.smiles):
    fp.append(AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(m), 2,nBits=1024))

x = rdkit_numpy_convert(fp) # matrix of descriptors

# all_data['smiles'] = all_data['smiles'].apply(normalize_smile)
# all_data['enc_smiles'] = all_data['smiles'].apply(encode_smile)
descr = []
# calc some descriptors
for m in tqdm(all_data.smiles):
    descr.append([Descriptors.MolLogP(Chem.MolFromSmiles(m)),
                  Descriptors.TPSA(Chem.MolFromSmiles(m)),
                  Descriptors.NHOHCount(Chem.MolFromSmiles(m)),
                  Descriptors.NOCount(Chem.MolFromSmiles(m)),
                  Descriptors.NumHAcceptors(Chem.MolFromSmiles(m)),
                  Descriptors.NumHDonors(Chem.MolFromSmiles(m)),
                  Descriptors.NumRotatableBonds(Chem.MolFromSmiles(m)),
                  Descriptors.NumHeteroatoms(Chem.MolFromSmiles(m)),
                  Descriptors.FractionCSP3(Chem.MolFromSmiles(m)),
                  Descriptors.ExactMolWt(Chem.MolFromSmiles(m)),
                  rdMolDescriptors.CalcNumAromaticRings(Chem.MolFromSmiles(m))])

descr = np.asarray(descr)
```

```python
cat_boost_models_dir = 'catboost_sol/'
os.makedirs(cat_boost_models_dir,exist_ok=True)
```

```python
indep_idx = []
with open(f'solubility_indep_ids.txt', 'r') as fp:
    for line in fp.read().splitlines():
        indep_idx.append(line)
```

```python
from torcheval.metrics import MulticlassAUPRC
from scipy.optimize import minimize
bounds = [(0.01,4)]*3
x0 = [1.0]*3
def kappa_val_func(x,y_train_answers,pred_train_probas):
    x1,x2,x3 = x
    local_preds = pred_train_probas**np.array([x1,x2,x3])
    return -round(cohen_kappa_score(y_train_answers, local_preds.argmax(axis=1),weights='quadratic'),4)
```

```python
for fold in range(5):
    val_ID = []
    with open(f'solubility_valid_id_{fold}.txt', 'r') as fp:
        for line in fp.read().splitlines():
            val_ID.append(line)

    train_ID = []
    with open(f'solubility_train_id_{fold}.txt', 'r') as fp:
        for line in fp.read().splitlines():
            train_ID.append(line)

    train_idx = all_data[all_data.Id.isin(train_ID)].index
    val_idx = all_data[all_data.Id.isin(val_ID)].index

    y_train = all_data[all_data.index.isin(train_idx)].sol_category.to_numpy()
    X_cat_train = features_df[features_df.index.isin(train_idx)].to_numpy()

    y_val = all_data[all_data.index.isin(val_idx)].sol_category.to_numpy()
    X_cat_val = features_df[features_df.index.isin(val_idx)].to_numpy()
    
    print(fold, len(X_cat_train),len(X_cat_val),len(y_train),len(y_val))

    cat_boost = CatBoostClassifier(learning_rate=0.1,depth=10,l2_leaf_reg=1,loss_function='MultiClass',
                                   task_type='GPU', od_wait=400, verbose=20, iterations=5000)
    
    cat_boost.fit(X=X_cat_train,y=y_train,eval_set=(X_cat_val,y_val), use_best_model=True)

    curr_name = f'catboost_1024_morgan_rdkit_split_{fold}'
    
    cat_boost.save_model(f'{cat_boost_models_dir}{curr_name}',
               format="cbm",
               export_parameters=None,
               pool=None)

    sol_types = ['LOW','MEDIUM','HIGH']


    ##kappa coefs
    pred_probas = cat_boost.predict_proba(X_cat_train)
    minim_res = minimize(kappa_val_func, x0,args=(y_train,pred_probas), method='powell',bounds=bounds)
    coefs = np.array(minim_res.x)

    with open(f'kappa_coefs_{curr_name}.txt', 'w') as fp:
        for item in coefs:
            # write each item on a new line
            fp.write("%s\n" % item)
    
    ##val
    pred_probas = cat_boost.predict_proba(X_cat_val)
    preds = cat_boost.predict(X_cat_val)
    precs = []
    recs = []
    aps = []

    
    
    for cl in range(0, 3):
        precision, recall, _ = precision_recall_curve((y_val==cl).astype(int), pred_probas[:, cl], pos_label=1)
        average_precision = average_precision_score((y_val==cl).astype(int), pred_probas[:, cl], pos_label=1)
        precs.append(precision)
        recs.append(recall)
        aps.append(average_precision)
    
    print(classification_report(y_val, preds, target_names=sol_types),
          file=open(cat_boost_models_dir+'validation_classification_report'+curr_name+'.txt', 'a'))
    print(classification_report(y_val, (pred_probas**coefs).argmax(axis=1), target_names=sol_types),
          file=open(cat_boost_models_dir+'coef_valdiation_classification_report'+curr_name+'.txt', 'a'))
    
    mpl.rcParams['figure.dpi'] = 150
    kappa = round(cohen_kappa_score(y_val, preds,weights='quadratic'),4)
    adj_kappa = round(cohen_kappa_score(y_val, (pred_probas**coefs).argmax(axis=1),weights='quadratic'),4)
    plt.title(f'Precision-recall curve, Quadratic kappa {kappa}, adjusted {adj_kappa}', fontsize=14)
    
    for marker in range(3):
        plt.plot(recs[marker], precs[marker], label=f'AUC {sol_types[marker]} = %0.2f' % aps[marker])
    plt.legend(loc='lower right')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.xticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.yticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.grid(True)
    plt.savefig(cat_boost_models_dir+f'valdiation_precision_recall_{curr_name}.png', format="png")
    plt.close()


    ##indep
    pred_probas = cat_boost.predict_proba(features_df[features_df.index.isin(all_data[all_data.Id.isin(indep_idx)].index)].to_numpy())
    preds = cat_boost.predict(features_df[features_df.index.isin(all_data[all_data.Id.isin(indep_idx)].index)].to_numpy())
    y_ind = all_data[all_data.Id.isin(indep_idx)].sol_category.to_numpy()
    precs = []
    recs = []
    aps = []
    
    for cl in range(0, 3):
        precision, recall, _ = precision_recall_curve((y_ind==cl).astype(int), pred_probas[:, cl], pos_label=1)
        average_precision = average_precision_score((y_ind==cl).astype(int), pred_probas[:, cl], pos_label=1)
        precs.append(precision)
        recs.append(recall)
        aps.append(average_precision)
    
    print(classification_report(y_ind, preds, target_names=sol_types),
          file=open(cat_boost_models_dir+'indep_classification_report'+curr_name+'.txt', 'a'))
    print(classification_report(y_ind, (pred_probas**coefs).argmax(axis=1), target_names=sol_types),
          file=open(cat_boost_models_dir+'coef_indep_classification_report'+curr_name+'.txt', 'a'))
    
    
    mpl.rcParams['figure.dpi'] = 150
    kappa = round(cohen_kappa_score(y_ind, preds,weights='quadratic'),4)
    adj_kappa = round(cohen_kappa_score(y_ind, (pred_probas**coefs).argmax(axis=1),weights='quadratic'),4)
    plt.title(f'Precision-recall curve, Quadratic kappa {kappa}, adjusted {adj_kappa}', fontsize=14)
   
    for marker in range(3):
        plt.plot(recs[marker], precs[marker], label=f'AUC {sol_types[marker]} = %0.2f' % aps[marker])
    plt.legend(loc='lower right')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.xticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.yticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.grid(True)
    plt.savefig(cat_boost_models_dir+f'indep_precision_recall_{curr_name}.png', format="png")
    plt.close()
```

```python
cat_boost_models_dir = 'catboost_sol/'
os.makedirs(cat_boost_models_dir,exist_ok=True)
indep_idx = []
with open(f'solubility_indep_ids.txt', 'r') as fp:
    for line in fp.read().splitlines():
        indep_idx.append(line)
ind_id = all_data[all_data.Id.isin(indep_idx)].index
from torcheval.metrics import MulticlassAUPRC
from scipy.optimize import minimize
bounds = [(0.01,4)]*3
x0 = [1.0]*3
def kappa_val_func(x,y_train_answers,pred_train_probas):
    x1,x2,x3 = x
    local_preds = pred_train_probas**np.array([x1,x2,x3])
    return -round(cohen_kappa_score(y_train_answers, local_preds.argmax(axis=1),weights='quadratic'),4)
```

```python
val_ID = []
with open(f'solubility_valid_id_{0}.txt', 'r') as fp:
    for line in fp.read().splitlines():
        val_ID.append(line)

train_ID = []
with open(f'solubility_train_id_{0}.txt', 'r') as fp:
    for line in fp.read().splitlines():
        train_ID.append(line)

uni_id = val_ID+train_ID

train_idx = all_data[all_data.Id.isin(uni_id)].index
```

```python
all_probas = []
for fold in range(5):
    curr_name = f'catboost_1024_morgan_rdkit_split_{fold}'
    y_train = all_data[all_data.index.isin(train_idx)].sol_category.to_numpy()
    X_cat_train = features_df[features_df.index.isin(train_idx)].to_numpy()

    cat_boost = CatBoostClassifier(learning_rate=0.1,depth=10,l2_leaf_reg=1,loss_function='MultiClass',
                                   task_type='CPU', od_wait=400, verbose=20, iterations=5000)
    cat_boost.load_model(f'{cat_boost_models_dir}{curr_name}',
               format="cbm")

    sol_types = ['LOW','MEDIUM','HIGH']

    ##kappa coefs
    pred_probas = cat_boost.predict_proba(X_cat_train)
    all_probas.append(pred_probas)
```

```python
y_answers = y_train
pred_probas = np.array(all_probas).mean(axis=0)

from scipy.optimize import minimize
bounds = [(0.01,4)]*3
x0 = [1.0]*3
minim_res = minimize(kappa_val_func, x0,args=(y_answers,pred_probas), method='powell',bounds=bounds)
coefs = np.array(minim_res.x)

with open(f'kappa_coefs_5_catboost_together_{curr_name}.txt', 'w') as fp:
    for item in coefs:
        # write each item on a new line
        fp.write("%s\n" % item)
```

```python
all_probas_indep = []
for fold in range(5):
    curr_name = f'catboost_1024_morgan_rdkit_split_{fold}'
    
    y_indep = all_data[all_data.index.isin(ind_id)].sol_category.to_numpy()
    
    X_cat_indep = features_df[features_df.index.isin(ind_id)].to_numpy()

    cat_boost = CatBoostClassifier(learning_rate=0.1,depth=10,l2_leaf_reg=1,loss_function='MultiClass',
                                   task_type='CPU', od_wait=400, verbose=20, iterations=5000)
    cat_boost.load_model(f'{cat_boost_models_dir}{curr_name}',
               format="cbm")

    sol_types = ['LOW','MEDIUM','HIGH']

    ##kappa coefs
    pred_probas_indep = cat_boost.predict_proba(X_cat_indep)
    all_probas_indep.append(pred_probas_indep)
```

```python
y_answers = y_indep
pred_probas = np.array(all_probas_indep).mean(axis=0)
```

```python
##indep
precs = []
recs = []
aps = []

for cl in range(0, 3):
    precision, recall, _ = precision_recall_curve((y_answers==cl).astype(int), pred_probas[:, cl], pos_label=1)
    average_precision = average_precision_score((y_answers==cl).astype(int), pred_probas[:, cl], pos_label=1)
    precs.append(precision)
    recs.append(recall)
    aps.append(average_precision)

print(classification_report(y_answers, pred_probas.argmax(axis=1), target_names=sol_types),
      file=open(cat_boost_models_dir+'5_split_indep_classification_report'+curr_name+'.txt', 'a'))
print(classification_report(y_answers, (pred_probas**coefs).argmax(axis=1), target_names=sol_types),
      file=open(cat_boost_models_dir+'5_split_coef_indep_classification_report'+curr_name+'.txt', 'a'))


mpl.rcParams['figure.dpi'] = 150
kappa = round(cohen_kappa_score(y_answers, pred_probas.argmax(axis=1),weights='quadratic'),4)
adj_kappa = round(cohen_kappa_score(y_answers, (pred_probas**coefs).argmax(axis=1),weights='quadratic'),4)
plt.title(f'Precision-recall curve, Quadratic kappa {kappa}, adjusted {adj_kappa}', fontsize=14)

for marker in range(3):
    plt.plot(recs[marker], precs[marker], label=f'AUC {sol_types[marker]} = %0.2f' % aps[marker])
plt.legend(loc='lower right')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.xticks(np.arange(0, 1.1, 0.1), fontsize=9)
plt.yticks(np.arange(0, 1.1, 0.1), fontsize=9)
plt.grid(True)
plt.savefig(cat_boost_models_dir+f'indep_precision_recall_5_split_{curr_name}.png', format="png")
plt.close()
```
