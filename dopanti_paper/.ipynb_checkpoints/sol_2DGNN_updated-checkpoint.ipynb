```python
ATMS_DICT = {'P': 0,
 'F': 1,
 'N': 2,
 'I': 3,
 'K': 4,
 'B': 5,
 'O': 6,
 'C': 7,
 'Cl': 8,
 'S': 9,
 'Na': 10,
 'Br': 11}

HYBR_DICT = {'SP3D2': 0, 'SP3': 1, 'SP2': 2, 'SP': 3, 'S': 4}
```

```python
scaler_descr = torch.load('rdkit_descriptor_scaler_sol')
```

```python
def minmaxscaler(X, xmax,xmin):
    min = 0
    max = 1
    X_std = (X - xmin) / (xmax - xmin)
    return  X_std * (max - min) + min

def synth_lig_graph_collection(smile,y):
    
    mol = wrapped_mol_from_smiles(smile)

    Chem.SanitizeMol(mol)

    x_features = []
    atm_features = []
    hyb_features = []
    bonds = []
    bond_type_features = []
    bond_features = []

    bond_dict = {'SINGLE':0, 'AROMATIC':1, 'DOUBLE':2, 'TRIPLE':3}

    for i in range(mol.GetNumAtoms()):
        atom = mol.GetAtomWithIdx(i)
        atm_features.append([ATMS_DICT[atom.GetSymbol()]])
        hyb_features.append([HYBR_DICT[str(atom.GetHybridization())]])
        x_features.append([minmaxscaler(int(atom.GetDegree()),5,0)] +
                       [minmaxscaler(int(atom.GetTotalNumHs()),5,0)] +
                        [minmaxscaler(int(atom.GetImplicitValence()),5,0)] +
                     [int(atom.IsInRing())]+
                         [int(atom.GetIsAromatic())])

    for bond in mol.GetBonds():
        bond_type_features.append(bond_dict[str(bond.GetBondType())])
        bond_type_features.append(bond_dict[str(bond.GetBondType())])
        
        feature = [[int(bond.IsInRing())]+
                             [int(bond.GetIsConjugated())]]
        bond_features.append(feature)
        
        bond_features.append(feature)
        
        bnd = [bond.GetBeginAtomIdx(),bond.GetEndAtomIdx()]
        bonds.append(bnd)
        bonds.append(list(reversed(bnd)))

    rdkit_features = scaler_descr.transform(np.array([[Descriptors.MolLogP(Chem.MolFromSmiles(smile)),
                  Descriptors.TPSA(Chem.MolFromSmiles(smile)),
                  Descriptors.NHOHCount(Chem.MolFromSmiles(smile)),
                  Descriptors.NOCount(Chem.MolFromSmiles(smile)),
                  Descriptors.NumHAcceptors(Chem.MolFromSmiles(smile)),
                  Descriptors.NumHDonors(Chem.MolFromSmiles(smile)),
                  Descriptors.NumRotatableBonds(Chem.MolFromSmiles(smile)),
                  Descriptors.NumHeteroatoms(Chem.MolFromSmiles(smile)),
                  Descriptors.FractionCSP3(Chem.MolFromSmiles(smile)),
                  Descriptors.ExactMolWt(Chem.MolFromSmiles(smile)),
                  rdMolDescriptors.CalcNumAromaticRings(Chem.MolFromSmiles(smile))]]))
        
    data = Data(x=torch.from_numpy(np.vstack(x_features)), 
                atm_feat = torch.tensor(np.array(atm_features), dtype=torch.long),
                hyb_feat = torch.tensor(np.array(hyb_features), dtype=torch.long),
                rdkit = torch.tensor(rdkit_features),
                edge_index=torch.tensor(np.array(bonds), dtype=torch.long).t().contiguous(),
                edge_attr=torch.tensor(np.vstack(bond_features)),
                edge_emb_attr = torch.tensor(np.vstack(bond_type_features), dtype=torch.long),
               y = torch.tensor([y],dtype=torch.long))

    return data
```

```python
# V1N
import torch.nn.functional as F
from torch.nn import Linear, Conv1d, Embedding
from torch_geometric.nn import BatchNorm
from torch_geometric.nn import global_mean_pool,global_max_pool,global_add_pool,BatchNorm, TransformerConv


class GNN(torch.nn.Module):
    def __init__(self,
                 neurons=128,
                 edge_dim=16,
                 linear_neurons=1024,
                 num_heads=16):

        super(GNN, self).__init__()

        self.neurons = neurons

        self.actfc = nn.LeakyReLU(negative_slope=0.2)
        
        self.atm_emb = Embedding(num_embeddings=12, embedding_dim=16)
        self.hybr_emb = Embedding(num_embeddings=5, embedding_dim=16)
        self.bond_emb = Embedding(num_embeddings=4, embedding_dim=8)

        self.edge_mlp = nn.Sequential(Linear(in_features=10, out_features=16),
                                      self.actfc,
                                      nn.BatchNorm1d(16),
                                      Linear(in_features=16, out_features=16),
                                      self.actfc)

        self.conv_layers = nn.ModuleList()

        self.conv_sequentials = nn.ModuleList()

        self.batchnorm = nn.BatchNorm1d(neurons*3)

        self.layer_1 = TransformerConv(37, neurons // num_heads, heads=num_heads, concat=True, dropout=0,
                                edge_dim=edge_dim)

        for i in range(6):
            self.conv_layers.append(
                TransformerConv(neurons, neurons // num_heads, heads=num_heads, concat=True, dropout=0.2,
                                edge_dim=edge_dim))
            self.conv_sequentials.append(nn.Sequential(
                nn.BatchNorm1d(neurons),
                self.actfc,
            ))

        self.out_mlp = nn.Sequential(Linear(in_features=(neurons*3)+11, out_features=linear_neurons),
                                     nn.BatchNorm1d(linear_neurons),
                                     self.actfc,
                                     nn.Dropout(0.5),
                                     Linear(in_features=linear_neurons, out_features=linear_neurons // 2),
                                     nn.BatchNorm1d(linear_neurons // 2),
                                     self.actfc,
                                     nn.Dropout(0.5),
                                     Linear(in_features=linear_neurons // 2, out_features=linear_neurons // 2),
                                     nn.BatchNorm1d(linear_neurons // 2),
                                     self.actfc,
                                     nn.Dropout(0.5))

        self.layer_out = Linear(in_features=linear_neurons // 2, out_features=3)

    def forward(self, x, atm_feat, hyb_feat, edge_index, batch_index, edge_attr, edge_emb_attr, rdkit):

        x = torch.cat((x.float(), self.atm_emb(atm_feat).reshape(-1,16), self.hybr_emb(hyb_feat).reshape(-1,16)), dim=1)

        edge_attr = self.edge_mlp(torch.cat((edge_attr.float(), self.bond_emb(edge_emb_attr).reshape(-1,8)), dim=1))

        out = self.layer_1(x, edge_index, edge_attr)

        for i in range(6):
            out = self.conv_layers[i](out, edge_index, edge_attr)
            out = self.conv_sequentials[i](out)  

        out_mean = global_mean_pool(out, batch_index)

        out_max = global_mean_pool(out, batch_index)

        out_add = global_add_pool(out, batch_index)

        out = torch.cat((out_mean, out_max, out_add), dim=1)

        out = self.batchnorm(out)

        out = torch.cat((out,rdkit.float()), dim=1)

        out = self.out_mlp(out)
        
        out = self.layer_out(out)

        return out
```

```python
def train_epoch(model, optimizer, train_loader, criterion, device):
    model.train()
    for data in train_loader:
        data = data.to(device)  
        out = model(data.x, data.atm_feat, data.hyb_feat, data.edge_index, data.batch, data.edge_attr, data.edge_emb_attr, data.rdkit)
        loss = criterion(out, data.y)

        # backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

def evaluate_loss(loader, model, criterion, device, metric=None):
    model.eval()
    with torch.no_grad():
        loss = 0
        j = 0
        sg = nn.Softmax()
        for data in loader:
            data = data.to(device)  
            out = model(data.x, data.atm_feat, data.hyb_feat, data.edge_index, data.batch, data.edge_attr, data.edge_emb_attr, data.rdkit)
            loss += criterion(out, data.y)
            if metric is not None:
                metric.update(sg(out), data.y.long())
            j += 1
        return (loss / j)


def train_m_v3(model, opt, train_loader, val_loader, criterion, n_epochs, device, name, 
               metric=None, early_stop=5,scheduler=None, verbose=True):
    train_log = []
    val_log = []
    val_metric_log = []
    train_metric_log = []
    val_min = np.inf
    train_min = np.inf
    epoch_min = 0
    fixed_epoch_val = 0
    fixed_epoch_train = 0
    val_allowed_maximum = np.inf
    early_counter = 0

    for epoch in tqdm(range(n_epochs)):
        train_epoch(model, opt, train_loader, criterion, device)
        train_loss = evaluate_loss(train_loader, model, criterion, device, metric)
        
        if metric is not None:
            train_metric = metric.compute().cpu().numpy().item()
            metric.reset()
        elif metric is None:
            train_metric = torch.tensor([0])

        val_loss = evaluate_loss(val_loader, model,
                                 criterion, device, metric)
        if metric is not None:
            val_metric = metric.compute().cpu().numpy().item()
            metric.reset()

        elif metric is None:
            train_metric = torch.tensor([0])
        
        if scheduler is not None:
            scheduler.step(val_loss)
            
        train_log.append(train_loss.cpu().numpy().item())
        val_log.append(val_loss.cpu().numpy().item())

        train_metric_log.append(train_metric)
        val_metric_log.append(val_metric)
        
        if val_loss < val_min:
            val_min = val_loss.cpu().numpy().item()
            epoch_val_absolute_min = epoch
            val_allowed_maximum = val_loss * 1.0001 #removed allowance for validation fluctuation

        if train_loss < train_min:
            train_min = train_loss

        if val_loss <= val_allowed_maximum:
            epoch_min = epoch
            fixed_epoch_val = val_loss.cpu().numpy().item()
            fixed_epoch_train = train_min.cpu().numpy().item()
            torch.save(model.state_dict(), name)
            torch.save({
                'epoch': epoch_min,
                'optimizer_state_dict': optimizer.state_dict(),
                'train_loss_history': train_log,
                'val_loss_history': val_log,
                'train_metric_history': train_metric_log,
                'val_metric_history': val_metric_log,
            }, name + 'training_optimizer_logs_')
            early_counter = 0

        if val_loss > val_allowed_maximum:
            early_counter += 1

        if verbose:
            # if epoch % 1 == 0:
            print(f'Epoch [{epoch + 1}/{n_epochs}], Loss (train/test): {train_loss}/{val_loss},'  
            f'AP (train/test): {train_metric_log[-1]}/{val_metric_log[-1]}')

        res_dict = {'min_loss_epoch': epoch_min,
                    'train_min': train_min.cpu().numpy().item(),
                    'train_loss_history': train_log,
                    'val_loss_history': val_log,
                    'val_min': val_min,
                    'final_val': fixed_epoch_val,
                    'final_train': fixed_epoch_train,
                'train_metric_history': train_metric_log,
                'val_metric_history': val_metric_log}

        if early_counter == early_stop:
            print(
                f'Early stopping after epoch {epoch}, model saved at epoch {epoch_min}, validation loss {fixed_epoch_val}.')
            return res_dict

    return res_dict
```

```python
def kappa_val_func(x,y_train_answers,pred_train_probas):
    x1,x2,x3 = x
    local_preds = pred_train_probas**np.array([x1,x2,x3])
    return -round(cohen_kappa_score(y_train_answers, local_preds.argmax(axis=1),weights='quadratic'),4)
```

```python
from torcheval.metrics import MulticlassAUPRC
from scipy.optimize import minimize
bounds = [(0.01,4)]*3
x0 = [1.0]*3
##V1 - BATCH_SIZE = int(2048) EPOCHS = 60 LR = 1e-4 WD = 0.00
## A version - BATCH_SIZE = int(2048) EPOCHS = 60 LR = 1e-3 WD = 0.00
TARGET = ['sol_category']
MODEL_VERSION = 'V1_A'
EARLY_STOP = 10
BATCH_SIZE = int(2048)
LR = 1e-3
WD = 0.00
EPOCHS = 100
WH_BALANCE = ''
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

indep_loader = torch_geometric.loader.DataLoader(list(all_data['graphs'].loc[ind_id]), batch_size=BATCH_SIZE, shuffle=False)

for fold in range(0,1):
    val_ID = []
    with open(f'solubility_valid_id_{fold}.txt', 'r') as fp:
        for line in fp.read().splitlines():
            val_ID.append(line)

    train_ID = []
    with open(f'solubility_train_id_{fold}.txt', 'r') as fp:
        for line in fp.read().splitlines():
            train_ID.append(line)

    train_idx = all_data[all_data.Id.isin(train_ID)].index
    val_idx = all_data[all_data.Id.isin(val_ID)].index
    
    train_loader = torch_geometric.loader.DataLoader(list(all_data['graphs'].loc[train_idx]), batch_size=BATCH_SIZE, shuffle=True)

    val_loader = torch_geometric.loader.DataLoader(list(all_data['graphs'].loc[val_idx]), batch_size=BATCH_SIZE, shuffle=False)
    
    curr_name = f"GNN_SOL_{fold}_{MODEL_VERSION}_{WH_BALANCE}"
    
    neural = GNN()
    
    # neural = nn.DataParallel(neural)
    neural = neural.to(device)

    metric = MulticlassAUPRC(num_classes=3)
    metric.to(device)
    # optimizer = AdamW(neural.parameters(), lr=LR, weight_decay=WD)
    optimizer = Adam(neural.parameters(), lr=LR)
    # scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.1, patience=5)
    class_weights=class_weight.compute_class_weight(class_weight='balanced',classes=np.unique(all_data[TARGET].loc[train_idx, TARGET].to_numpy()),y=all_data[TARGET].loc[train_idx, TARGET].to_numpy().flatten())
    class_weights=torch.tensor(class_weights,dtype=torch.float).to(device)
    if WH_BALANCE == 'balanced':
        loss = nn.CrossEntropyLoss(weight=class_weights)
    else:
        loss = nn.CrossEntropyLoss()

    neural_folder = 'GNN_pytorch_neurals/'

    os.makedirs(neural_folder,exist_ok=True)
    
    res_dict = train_m_v3(neural, optimizer, train_loader, val_loader, loss, EPOCHS, 
                          device, neural_folder+curr_name,metric,early_stop=EARLY_STOP, verbose=True) 
        
    with torch.cuda.device(device):
        torch.cuda.empty_cache()
    
    mpl.rcParams['figure.dpi'] = 150
    print(res_dict['min_loss_epoch'], res_dict['final_val'], res_dict['final_train'], res_dict['val_min'])
    
    with open(os.path.join(neural_folder, curr_name+'_results.txt'), "w") as f:
        f.write(json.dumps(res_dict))

    y1 = []
    y2 = []
    for lg in res_dict['train_loss_history']:
        y1.append(lg)
    for lg in res_dict['val_loss_history']:
        y2.append(lg)
    x = range(1,len(y1)+1)
    plt.plot(x,y1, label = 'train loss')
    plt.plot(x,y2, label = 'val loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig(neural_folder+curr_name+'_loss_func.png')
    plt.clf()
    plt.close()
    
    sol_types = ['LOW','MEDIUM','HIGH']
    
    ##find coefs
    neural = GNN()
    neural.load_state_dict(torch.load(neural_folder+curr_name,map_location=device))
    neural = neural.to(device)
    neural.eval()
    
    y_answers = []
    pred_probas = []
    with torch.no_grad():
        m = nn.Softmax(dim=1)
        for data in train_loader:
            data = data.to(device)  
            out = neural(data.x, data.atm_feat, data.hyb_feat, data.edge_index, data.batch, data.edge_attr, data.edge_emb_attr, data.rdkit)
            pred = m(out.to(torch.device("cpu"))).detach().numpy()
            y_answers.extend(data.y.to(torch.device("cpu")).detach().numpy())
            pred_probas.extend(pred)
    y_answers = np.array(y_answers)
    pred_probas = np.array(pred_probas)

    minim_res = minimize(kappa_val_func, x0,args=(y_answers,pred_probas), method='powell',bounds=bounds)
    coefs = np.array(minim_res.x)

    with open(f'kappa_coefs_{curr_name}.txt', 'w') as fp:
        for item in coefs:
            # write each item on a new line
            fp.write("%s\n" % item)
        
    ##val
    y_answers = []
    pred_probas = []
    with torch.no_grad():
        m = nn.Softmax(dim=1)
        for data in val_loader:
            data = data.to(device)  
            out = neural(data.x, data.atm_feat, data.hyb_feat, data.edge_index, data.batch, data.edge_attr, data.edge_emb_attr, data.rdkit)
            pred = m(out.to(torch.device("cpu"))).detach().numpy()
            y_answers.extend(data.y.to(torch.device("cpu")).detach().numpy())
            pred_probas.extend(pred)
    y_answers = np.array(y_answers)
    pred_probas = np.array(pred_probas)
    
    precs = []
    recs = []
    aps = []
    
    for cl in range(0, 3):
        precision, recall, _ = precision_recall_curve((y_answers==cl).astype(int), pred_probas[:, cl], pos_label=1)
        average_precision = average_precision_score((y_answers==cl).astype(int), pred_probas[:, cl], pos_label=1)
        precs.append(precision)
        recs.append(recall)
        aps.append(average_precision)
    
    print(classification_report(y_answers, pred_probas.argmax(axis=1), target_names=sol_types),
          file=open(neural_folder+'valdiation_classification_report'+curr_name+'.txt', 'a'))

    print(classification_report(y_answers, (pred_probas**coefs).argmax(axis=1), target_names=sol_types),
          file=open(neural_folder+'coef_valdiation_classification_report'+curr_name+'.txt', 'a'))
    
    
    mpl.rcParams['figure.dpi'] = 150
    kappa = round(cohen_kappa_score(y_answers, pred_probas.argmax(axis=1),weights='quadratic'),4)
    adj_kappa = round(cohen_kappa_score(y_answers, (pred_probas**coefs).argmax(axis=1),weights='quadratic'),4)
    plt.title(f'Precision-recall curve, Quadratic kappa {kappa}, adjusted {adj_kappa}', fontsize=14)
    
    for marker in range(3):
        plt.plot(recs[marker], precs[marker], label=f'AUC {sol_types[marker]} = %0.2f' % aps[marker])
    plt.legend(loc='lower right')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.xticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.yticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.grid(True)
    plt.savefig(neural_folder+f'valdiation_precision_recall_{curr_name}.png', format="png")
    plt.close()

    ##indep
    y_answers = []
    pred_probas = []
    with torch.no_grad():
        m = nn.Softmax(dim=1)
        for data in indep_loader:
            data = data.to(device)  
            out = neural(data.x, data.atm_feat, data.hyb_feat, data.edge_index, data.batch, data.edge_attr, data.edge_emb_attr, data.rdkit)
            pred = m(out.to(torch.device("cpu"))).detach().numpy()
            y_answers.extend(data.y.to(torch.device("cpu")).detach().numpy())
            pred_probas.extend(pred)
    y_answers = np.array(y_answers)
    pred_probas = np.array(pred_probas)
    precs = []
    recs = []
    aps = []
    
    for cl in range(0, 3):
        precision, recall, _ = precision_recall_curve((y_answers==cl).astype(int), pred_probas[:, cl], pos_label=1)
        average_precision = average_precision_score((y_answers==cl).astype(int), pred_probas[:, cl], pos_label=1)
        precs.append(precision)
        recs.append(recall)
        aps.append(average_precision)
    
    print(classification_report(y_answers, pred_probas.argmax(axis=1), target_names=sol_types),
          file=open(neural_folder+'indep_classification_report'+curr_name+'.txt', 'a'))
    print(classification_report(y_answers, (pred_probas**coefs).argmax(axis=1), target_names=sol_types),
          file=open(neural_folder+'coef_indep_classification_report'+curr_name+'.txt', 'a'))
    
    
    mpl.rcParams['figure.dpi'] = 150
    kappa = round(cohen_kappa_score(y_answers, pred_probas.argmax(axis=1),weights='quadratic'),4)
    adj_kappa = round(cohen_kappa_score(y_answers, (pred_probas**coefs).argmax(axis=1),weights='quadratic'),4)
    plt.title(f'Precision-recall curve, Quadratic kappa {kappa}, adjusted {adj_kappa}', fontsize=14)
    
    for marker in range(3):
        plt.plot(recs[marker], precs[marker], label=f'AUC {sol_types[marker]} = %0.2f' % aps[marker])
    plt.legend(loc='lower right')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.xticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.yticks(np.arange(0, 1.1, 0.1), fontsize=9)
    plt.grid(True)
    plt.savefig(neural_folder+f'indep_precision_recall_{curr_name}.png', format="png")
    plt.close()
```

```python
# V1N
import torch.nn.functional as F
from torch.nn import Linear, Conv1d, Embedding
from torch_geometric.nn import BatchNorm
from torch_geometric.nn import global_mean_pool,global_max_pool,global_add_pool,BatchNorm, TransformerConv


class GNN(torch.nn.Module):
    def __init__(self,
                 neurons=128,
                 edge_dim=16,
                 linear_neurons=1024,
                 num_heads=16):

        super(GNN, self).__init__()

        self.neurons = neurons

        self.actfc = nn.LeakyReLU(negative_slope=0.2)
        
        self.atm_emb = Embedding(num_embeddings=12, embedding_dim=16)
        self.hybr_emb = Embedding(num_embeddings=5, embedding_dim=16)
        self.bond_emb = Embedding(num_embeddings=4, embedding_dim=8)

        self.edge_mlp = nn.Sequential(Linear(in_features=10, out_features=16),
                                      self.actfc,
                                      nn.BatchNorm1d(16),
                                      Linear(in_features=16, out_features=16),
                                      self.actfc)

        self.conv_layers = nn.ModuleList()

        self.conv_sequentials = nn.ModuleList()

        self.batchnorm = nn.BatchNorm1d(neurons*3)

        self.layer_1 = TransformerConv(37, neurons // num_heads, heads=num_heads, concat=True, dropout=0,
                                edge_dim=edge_dim)

        for i in range(6):
            self.conv_layers.append(
                TransformerConv(neurons, neurons // num_heads, heads=num_heads, concat=True, dropout=0.2,
                                edge_dim=edge_dim))
            self.conv_sequentials.append(nn.Sequential(
                nn.BatchNorm1d(neurons),
                self.actfc,
            ))

        self.out_mlp = nn.Sequential(Linear(in_features=(neurons*3)+11, out_features=linear_neurons),
                                     nn.BatchNorm1d(linear_neurons),
                                     self.actfc,
                                     nn.Dropout(0.5),
                                     Linear(in_features=linear_neurons, out_features=linear_neurons // 2),
                                     nn.BatchNorm1d(linear_neurons // 2),
                                     self.actfc,
                                     nn.Dropout(0.5),
                                     Linear(in_features=linear_neurons // 2, out_features=linear_neurons // 2),
                                     nn.BatchNorm1d(linear_neurons // 2),
                                     self.actfc,
                                     nn.Dropout(0.5))

        self.layer_out = Linear(in_features=linear_neurons // 2, out_features=3)

    def forward(self, x, atm_feat, hyb_feat, edge_index, batch_index, edge_attr, edge_emb_attr, rdkit):

        x = torch.cat((x.float(), self.atm_emb(atm_feat).reshape(-1,16), self.hybr_emb(hyb_feat).reshape(-1,16)), dim=1)

        edge_attr = self.edge_mlp(torch.cat((edge_attr.float(), self.bond_emb(edge_emb_attr).reshape(-1,8)), dim=1))

        out = self.layer_1(x, edge_index, edge_attr)

        for i in range(6):
            out = self.conv_layers[i](out, edge_index, edge_attr)
            out = self.conv_sequentials[i](out)  

        out_mean = global_mean_pool(out, batch_index)

        out_max = global_mean_pool(out, batch_index)

        out_add = global_add_pool(out, batch_index)

        out = torch.cat((out_mean, out_max, out_add), dim=1)

        out = self.batchnorm(out)

        out = torch.cat((out,rdkit.float()), dim=1)

        out = self.out_mlp(out)
        
        out = self.layer_out(out)

        return out
```

```python
def kappa_val_func(x,y_train_answers,pred_train_probas):
    x1,x2,x3 = x
    local_preds = pred_train_probas**np.array([x1,x2,x3])
    return -round(cohen_kappa_score(y_train_answers, local_preds.argmax(axis=1),weights='quadratic'),4)
```

```python
def minmaxscaler(X, xmax,xmin):
    min = 0
    max = 1
    X_std = (X - xmin) / (xmax - xmin)
    return  X_std * (max - min) + min

def synth_lig_graph_collection_no_y(smile):
    
    mol = wrapped_mol_from_smiles(smile)

    Chem.SanitizeMol(mol)

    x_features = []
    atm_features = []
    hyb_features = []
    bonds = []
    bond_type_features = []
    bond_features = []

    bond_dict = {'SINGLE':0, 'AROMATIC':1, 'DOUBLE':2, 'TRIPLE':3}

    for i in range(mol.GetNumAtoms()):
        atom = mol.GetAtomWithIdx(i)
        try:
            atm_features.append([ATMS_DICT[atom.GetSymbol()]])
        except KeyError:
            atm_features.append([ATMS_DICT['K']])
        hyb_features.append([HYBR_DICT[str(atom.GetHybridization())]])
        x_features.append([minmaxscaler(int(atom.GetDegree()),5,0)] +
                       [minmaxscaler(int(atom.GetTotalNumHs()),5,0)] +
                        [minmaxscaler(int(atom.GetImplicitValence()),5,0)] +
                     [int(atom.IsInRing())]+
                         [int(atom.GetIsAromatic())])

    for bond in mol.GetBonds():
        bond_type_features.append(bond_dict[str(bond.GetBondType())])
        bond_type_features.append(bond_dict[str(bond.GetBondType())])
        
        feature = [[int(bond.IsInRing())]+
                             [int(bond.GetIsConjugated())]]
        bond_features.append(feature)
        
        bond_features.append(feature)
        
        bnd = [bond.GetBeginAtomIdx(),bond.GetEndAtomIdx()]
        bonds.append(bnd)
        bonds.append(list(reversed(bnd)))

    rdkit_features = scaler_descr.transform(np.array([[Descriptors.MolLogP(Chem.MolFromSmiles(smile)),
                  Descriptors.TPSA(Chem.MolFromSmiles(smile)),
                  Descriptors.NHOHCount(Chem.MolFromSmiles(smile)),
                  Descriptors.NOCount(Chem.MolFromSmiles(smile)),
                  Descriptors.NumHAcceptors(Chem.MolFromSmiles(smile)),
                  Descriptors.NumHDonors(Chem.MolFromSmiles(smile)),
                  Descriptors.NumRotatableBonds(Chem.MolFromSmiles(smile)),
                  Descriptors.NumHeteroatoms(Chem.MolFromSmiles(smile)),
                  Descriptors.FractionCSP3(Chem.MolFromSmiles(smile)),
                  Descriptors.ExactMolWt(Chem.MolFromSmiles(smile)),
                  rdMolDescriptors.CalcNumAromaticRings(Chem.MolFromSmiles(smile))]]))
        
    data = Data(x=torch.from_numpy(np.vstack(x_features)), 
                atm_feat = torch.tensor(np.array(atm_features), dtype=torch.long),
                hyb_feat = torch.tensor(np.array(hyb_features), dtype=torch.long),
                rdkit = torch.tensor(rdkit_features),
                edge_index=torch.tensor(np.array(bonds), dtype=torch.long).t().contiguous(),
                edge_attr=torch.tensor(np.vstack(bond_features)),
                edge_emb_attr = torch.tensor(np.vstack(bond_type_features), dtype=torch.long))

    return data
```
